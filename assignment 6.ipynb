```python
df1 = pd.DataFrame({
    'ID':[1,2,3,4],
    'Name':['A','B','C','D']
})

df2 = pd.DataFrame({
    'ID':[2,3,4,5],
    'City':['Pune','Mumbai','Delhi','Chennai']
})
print(df1)
print(df2)
```

```python
pd.merge(df1, df2, on='ID', how='inner')
```

```python
pd.merge(df1, df2, on='ID', how='left')
```

```python
pd.merge(df1, df2, on='ID', how='right')
```

```python
pd.merge(df1, df2, on='ID', how='outer')
```

```python
pd.merge(df1, df2, on='ID', how='outer', indicator=True)\
.query("_merge != 'both'")
```

```python
pd.concat([df1, df2], axis=0)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
sales = pd.DataFrame({
    'Region':['East','West','East','North','West'],
    'Product':['A','A','B','B','A'],
    'Sales':[1000,1500,2000,1200,1800]
})
print(sales)
```

```python
pd.pivot_table(sales,
               values='Sales',
               index='Region',
               aggfunc='sum')
```

```python
pd.pivot_table(sales,
               values='Sales',
               index='Region',
               columns='Product',
               aggfunc='sum')
```

```python
pd.pivot_table(sales,
               values='Sales',
               index='Region',
               columns='Product',
               aggfunc='mean')
```

```python
data = pd.DataFrame({
    'Date':['2025-01-01','2025-01-02','2025-01-03',
            '2025-01-04','2025-01-05'],
    'Sales':[100,200,150,300,250]
})

# Question 18
data['Date'] = pd.to_datetime(data['Date'])

print(data)

data['Date'] = pd.to_datetime(data['Date'])

data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day

print(data)

```

```python
data.set_index('Date', inplace=True)

data['Rolling_Avg'] = data['Sales'].rolling(3).mean()

print(data)
```

```python
customers = pd.DataFrame({
    'CustomerID':[1,2,3],
    'CustomerName':['A','B','C']
})

orders = pd.DataFrame({
    'CustomerID':[1,2,1],
    'OrderAmount':[500,1000,700]
})

pd.merge(customers, orders, on='CustomerID')
```

```python
monthly = pd.DataFrame({
    'Month':['Jan','Jan','Feb','Feb'],
    'Product':['A','B','A','B'],
    'Sales':[1000,1200,1500,1800]
})

pd.pivot_table(monthly,
               values='Sales',
               index='Month',
               columns='Product',
               aggfunc='sum')
```

```python
daily = pd.DataFrame({
    'Date':pd.date_range('2025-01-01', periods=10),
    'Sales':[100,200,150,300,250,400,350,450,500,550]
})

daily.set_index('Date', inplace=True)

daily.resample('W').sum()
```

```python
daily['Cumulative_Sales'] = daily['Sales'].cumsum()

print(daily)
```
